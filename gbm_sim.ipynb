```python
import numpy as np
import matplotlib.pyplot as plt

# initial params
S0 = 100        # Initial stock price
mu = 0.05        # Drift
sigma = 0.2     # Volatility
T = 1.0         # Time horizon
dt = 0.01       # Time increment

N = int(T / dt)
t = np.linspace(0, T, N)

#  stock price array, BM
S = np.zeros(N)
B = np.zeros(N)
S[0] = S0

# BM increments
for i in range(1, N):
    B[i] = B[i-1] + np.random.normal(0, np.sqrt(dt)) 

# stock price using the SDE
for i in range(1, N):
    S[i] = S[i-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * (B[i] - B[i-1]))

plt.figure(figsize=(12, 6))
plt.plot(t, S, label='Stock Price (GBM)', color='blue')
plt.title('Stock Price Evolution with Adjusted Parameters')
plt.xlabel('Time (t)')
plt.ylabel('Stock Price (S_t)')
plt.grid()
plt.legend()
plt.show()
```
